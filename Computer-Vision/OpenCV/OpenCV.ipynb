# Image Handling

```python
#Import.
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```python
#cv2.imread(fileName, flag)

#Read Image.
image1 = cv2.imread("/content/IT Meme.PNG", cv2.IMREAD_COLOR) #IMREAD_COLOR = 1
image2 = cv2.imread("/content/IT Meme.PNG", cv2.IMREAD_GRAYSCALE) #IMREAD_GRAYSCALE = 0
image3 = cv2.imread("/content/IT Meme.PNG", cv2.IMREAD_UNCHANGED) #IMREAD_UNCHANGED = -1
```

```python
#cv2.imshow(title, image)
#cv2_imshow(image)

#cv2.waitkey(): 키보드 입력을 대기하는 함수, 0이면 입력할 때까지 대기한다.
#cv2.destroyAllWindows(): 화면에 나타난 윈도우를 종료한다.

#Print image1 & Shape.
print(image1.shape)
cv2_imshow(image1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Print image2 & Shape.
print(image2.shape)
cv2_imshow(image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Print image3 & Shape.
print(image3.shape)
cv2_imshow(image3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#cv2.imwrite(fileName, image)

cv2.imwrite("/content/image2.PNG", image2)
```

```python
#Check.
img_check = cv2.imread("/content/image2.PNG", 1)
print(img_check.shape)
cv2_imshow(img_check)
```

```python
#Using plt First.
img_plt = cv2.imread("/content/IT Meme.PNG", 1)

plt.imshow(img_plt)
plt.xticks([]) #X Axis Ticks.
plt.yticks([]) #Y Axis Ticks.
plt.show() 
```

```python
#Using plt Second.
#BGR to RGB.
img_plt = cv2.imread("/content/IT Meme.PNG", 1)

b, g, r = cv2.split(img_plt)
img_merge = cv2.merge([r, g, b])

plt.imshow(img_merge)
plt.xticks([]) #X Axis Ticks.
plt.yticks([]) #Y Axis Ticks.
plt.show() 
```
